```python
# SP_inv_S_gauss = np.eye(16*10)[:, :] - R_gauss
# SP_inv_S_RHS = np.eye(16*10)[:, :] - R_RHS
SP_inv_S_DHS = np.eye(16*10)[:, :] - R_DHS
SP_inv_S_DST = np.eye(16*10)[:, :] - R_DST
SP_inv_S_BHS = np.eye(16*10)[:, :] - R_BHS
SP_inv_S_BST = np.eye(16*10)[:, :] - R_BST

SP_inv_S_DHS_nodewise = np.eye(16*10)[:, :] - R_DHS_nodewise
SP_inv_S_DST_nodewise = np.eye(16*10)[:, :] - R_DST_nodewise
SP_inv_S_BHS_nodewise = np.eye(16*10)[:, :] - R_BHS_nodewise
SP_inv_S_BST_nodewise = np.eye(16*10)[:, :] - R_BST_nodewise
```

```python
# --- Traces as distributions (df_eff = tr(R) vs total shrinkage = tr(I-R)) ---
import matplotlib.pyplot as plt

tr_SPinvS_DHS   = np.trace(SP_inv_S_DHS,   axis1=1, axis2=2)
tr_SPinvS_DHS_nodewise   = np.trace(SP_inv_S_DHS_nodewise,   axis1=1, axis2=2)
tr_SPinvS_DST   = np.trace(SP_inv_S_DST,   axis1=1, axis2=2)
tr_SPinvS_DST_nodewise   = np.trace(SP_inv_S_DST_nodewise,   axis1=1, axis2=2)
tr_SPinvS_BHS   = np.trace(SP_inv_S_BHS,   axis1=1, axis2=2)
tr_SPinvS_BHS_nodewise   = np.trace(SP_inv_S_BHS_nodewise,   axis1=1, axis2=2)
tr_SPinvS_BST   = np.trace(SP_inv_S_BST,   axis1=1, axis2=2)
tr_SPinvS_BST_nodewise   = np.trace(SP_inv_S_BST_nodewise,   axis1=1, axis2=2)

```

```python

plt.figure(figsize=(8,4), dpi=150)
bins = 40
# plt.hist(tr_SPinvS_gauss, bins=bins, alpha=0.5, label="Gauss")
# plt.hist(tr_SPinvS_RHS,   bins=bins, alpha=0.5, label="RHS")
# plt.hist(tr_SPinvS_DHS,   bins=bins, alpha=0.5, label="DHS", color="C2")
plt.hist(tr_SPinvS_DHS_nodewise,   bins=bins, alpha=0.5, label="DHS - node", color="C2")
# plt.hist(tr_SPinvS_DST,   bins=bins, alpha=0.5, label="DST", color="C3")
plt.hist(tr_SPinvS_DST_nodewise,   bins=bins, alpha=0.5, label="DST - node", color="C3")
# plt.hist(tr_SPinvS_BHS,   bins=bins, alpha=0.5, label="BHS", color="C4")
plt.hist(tr_SPinvS_BHS_nodewise,   bins=bins, alpha=0.5, label="BHS - node", color="C4")
# plt.hist(tr_SPinvS_BST,   bins=bins, alpha=0.5, label="BST", color="C5")
plt.hist(tr_SPinvS_BST_nodewise,   bins=bins, alpha=0.5, label="BST - node", color="C5")
# plt.hist(tr_SPinvS_DST,   bins=bins, alpha=0.5, label="DST", color="C3")
# plt.hist(tr_SPinvS_BHS,   bins=bins, alpha=0.5, label="BHS", color="C4")
# plt.hist(tr_SPinvS_BST,   bins=bins, alpha=0.5, label="BST", color="C5")
plt.xlabel(r"$tr((P+S)^{-1}S)$", fontsize=15)
plt.ylabel("Frequency", fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(fontsize=15)
plt.tight_layout()
#plt.savefig("figures_for_use_in_paper/Friedman_m_eff_correlated_with_beta_no_gauss_or_RHS_correlated.pdf", bbox_inches="tight")
plt.show()
```
